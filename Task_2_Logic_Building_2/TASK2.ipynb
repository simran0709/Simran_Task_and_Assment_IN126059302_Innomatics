```python
#Problem 1: Unique Words in a Sentence
sentance="python is easy and python is powerful"
words=sentance.lower().split()
unique_words=set(words)
unique_words_count=len(unique_words)
print("unique words count:",unique_words_count)
print("unique words",unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary=0
highest_employee=""
for name,current_salary in employees.items():
  if current_salary>highest_salary:
    highest_salary=current_salary
    highest_employee=name
print("highest salary :",highest_employee,highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
  if price>1000:
    count +=1
print("the product above 1000:",count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# Count number of "P"
for status in attendance:
    if status == "P":
        present_count += 1

# Calculate percentage
percentage = (present_count / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers=set(phone_numbers)
print("unique_numbers",unique_numbers)
```

```python
#Problem 7: Count Character Frequency
sentance="python is easy language"
dic={}
for char in sentance:
  if char.isalpha():
    if char in dic:
      dic[char]+=1
    else:
      dic[char]=1
print(dic)
```
