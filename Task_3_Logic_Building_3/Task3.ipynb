```python
# Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score
highest_score = max(employees.values())

# Step 2: Collect employees who have the highest score
top_employees = []

for name, score in employees.items():
    if score == highest_score:
        top_employees.append(name)

# Step 3: Convert list to comma-separated string
names_string = ", ".join(top_employees)

# Step 4: Print final result
print(f"Top Performers Eligible for Bonus: {names_string} (Score: {highest_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

sentence = "Buy mobile phone buy phone online"

# Convert to lowercase
sentence = sentence.lower()

# Remove punctuation (if any)
import string
sentence = sentence.translate(str.maketrans('', '', string.punctuation))

# Count frequency
word_count = {}

for word in sentence.split():
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Display only words searched more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if reading is even
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domain and count
for email in emails:
    domain = email.split("@")[1]   # Get domain part
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average_sales = sum(sales) / len(sales)

# Detect spikes (30% above average)
threshold = average_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
