```python
def Smart_Parking(Logs, capacity):
    current_parking = 0
    peak_usage = 0

    # Process vehicle logs
    for log in Logs:
        if log == "IN":
            current_parking += 1
        elif log == "OUT":
            current_parking -= 1

        # Track peak usage
        if current_parking > peak_usage:
            peak_usage = current_parking

    # Check parking status AFTER processing all logs
    if current_parking >= capacity:
        parking_status = "Full"
    else:
        parking_status = "Available"

    print("Currently Parked Vehicles:", current_parking)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", parking_status)


# Sample Input
Logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
capacity = 50

Smart_Parking(Logs, capacity)
```

```python
#Problem Statement 2:Online Food Delivery Time Estimator

def Delivery_Time_Estimator(distance, traffic, weather):

    # Base time calculation (1 km = 5 minutes)
    base_time = distance * 5

    # Traffic delay
    if traffic == "Low":
        traffic_delay = 0
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 15
    else:
        traffic_delay = 0

    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 0
    elif weather == "Storm":
        weather_delay = 15
    else:
        weather_delay = 0

    # Final ETA
    final_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", final_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

Delivery_Time_Estimator(distance, traffic, weather)
```

```python
#Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def Movie_Analyzer(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 40:
        status = "Filling Fast"
    else:
        status = "Available"

    # Suggest additional show
    if occupancy >= 90:
        suggestion = "Open Additional Show"
    else:
        suggestion = "No Additional Show Needed"

    # Display results
    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

Movie_Analyzer(total_seats, booked_seats)
```

```python
#Problem 4: Cloud Server Load Classification System

def Server_Load_Classifier(cpu_readings):

    total = 0

    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {average_cpu:.0f}%")
    print("Server Status:", status)


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

Server_Load_Classifier(cpu_readings)
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor

def Classroom_Resource_Monitor(resource_usage):

    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

Classroom_Resource_Monitor(resource_usage)
```

```python
#Problem 6: Online Event Registration Capacity Controller

def Event_Registration_Controller(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

Event_Registration_Controller(capacity, registrations)
```
