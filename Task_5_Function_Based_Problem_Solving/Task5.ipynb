```python
#Problem 1: Social Media – Post Engagement Analyzer

def analyze_post(likes):
  total_likes = 0
  for like in likes:
    total_likes += like
  print("Total Likes:", total_likes)
  if total_likes>=1000:
      print("Post_Status:Viral Post")
  else:
     print("Post_Status:Normal Engagment")
likes = [300, 200, 150, 400]
analyze_post(likes)
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System

def check_stock_level(stocks):
  Medicine_stock=0
  for stock in stocks:
     Medicine_stock +=stock
  print("Medicine Stock:",Medicine_stock)
  if Medicine_stock<=10:
    print(" Status: Low Stock Alert")
  else:
    print("Status: Stock Sufficient")
stocks=[1,2,14]
check_stock_level(stocks)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
  total_rainful=0
  for rainfall in rainfall_data:
    total_rainful +=rainfall
  average = total_rainful / len(rainfall_data)
  print("Average Rainfall:",average)
  if average>=required_level:
    print("Rainfall Status:Adequate Rainfall")
  else:
    print("Rainfall Status:Inadequate Rainfall")
rainfall_data=[60, 80, 75, 70]
required_level=70
check_rainfall(rainfall_data, required_level)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
  total_usernames=len(usernames)
  print("Total Usernames:",total_usernames)
  duplicate_usernames=set(usernames)
  if len( duplicate_usernames)!= total_usernames:
    print("Duplicate Accounts Found:Yes",)
  else:
    print("Duplicate Accounts Found:No")
usernames=["simmu", "alex", "Nithin", "simmu"]
detect_duplicates(usernames)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def Student_Attendance_Eligibility_System(Attendances):
  Total_Classes=len(Attendances)
  Total_Present_Days=0
  for value in Attendances:
    if value==1:
      Total_Present_Days +=1
  Attendance_Percentage=(Total_Present_Days / Total_Classes)*100
  print("Attendance Percentage:",Attendance_Percentage)
  if Attendance_Percentage>=75:
    status = "Eligible"
  else:
    status = "Not Eligible"

  return status
Attendances=[0, 1, 1, 1, 0,0]
status=Student_Attendance_Eligibility_System(Attendances)
print("Attendance Status:", status)
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def Electricity_Bill_Analyzer(units):
    total_bill = 0

    # Slab Calculation
    if units <= 100:
        total_bill = units * 3

    elif units <= 200:
        total_bill = (100 * 3) + (units - 100) * 5

    else:
        total_bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage Classification
    if total_bill < 500:
        usage_status = "Low Usage"

    elif 500 <= total_bill <= 1500:
        usage_status = "Moderate Usage"

    else:
        usage_status = "High Usage"

    return total_bill, usage_status


# Example Usage
units = 220
bill, status = Electricity_Bill_Analyzer(units)

print("Units Consumed:", units)
print("Total Bill: ₹", bill)
print("Usage Level:", status)
```
